```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

# Timeseries

```python
filename = 'training_log_20171129_500batch_100epochs_timeseries_feedforward_model.csv'
csv_data = np.genfromtxt(filename, dtype=None, delimiter=',', names=True)
plt.figure()
plt.plot(csv_data['epoch'], csv_data['acc'], label='training')
plt.plot(csv_data['epoch'], csv_data['val_acc'], label='validation')
plt.legend(loc='best')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Timeries input + Dense model')
plt.ylim(0.5, 1)
```

```python
filename = 'training_log_20171129_500batch_100epochs_timeseries_convolutional_model.csv'
csv_data = np.genfromtxt(filename, dtype=None, delimiter=',', names=True)
plt.figure()
plt.plot(csv_data['epoch'], csv_data['acc'], label='training')
plt.plot(csv_data['epoch'], csv_data['val_acc'], label='validation')
plt.legend(loc='best')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Timeries input + Convolutional model')
plt.ylim(0.5, 1)
```
